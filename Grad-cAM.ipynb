This notebook visualize Grad-Cam activation from resnet models (only supporting resnet models for now) <br>
Implementation is available in Grad_Cam.py <br>
Credit: code is adpated from [here](http://snappishproductions.com/blog/2018/01/03/class-activation-mapping-in-pytorch.html)

```python
from Grad_Cam import *
import pandas as pd
import random
```

```python
def sample_img(seed):
    df = pd.read_csv("dataset.csv")
    test_df = df[df['type'] == "test"]
    sample_img = test_df.sample(n = 1,  random_state = seed)
    return sample_img['path'].values[0], sample_img['label'].values[0]
```

### Resnet 34

```python
seed = 10
test_img, test_label = sample_img(seed = seed)
plot_Cam(test_img, test_label, "res34")
```

```python
seed = 46
test_img, test_label = sample_img(seed = seed)
plot_Cam(test_img, test_label, "res34")
```

```python
seed = 80
test_img, test_label = sample_img(seed = seed)
plot_Cam(test_img, test_label, "res34")
```

```python
seed = 100
test_img, test_label = sample_img(seed = seed)
plot_Cam(test_img, test_label, "res34")
```

```python
seed = 200
test_img, test_label = sample_img(seed = seed)
plot_Cam(test_img, test_label, "res34")
```

```python
seed = 300
test_img, test_label = sample_img(seed = seed)
plot_Cam(test_img, test_label, "res34")
```
